# Data cleaning

```python
msno.matrix(df_hits)
```

```python
missing_values_hits = ((df_hits.isna().sum() / len(df_hits)) * 100).sort_values()
missing_values_sessions = ((df_sessions.isna().sum() / len(df_sessions)) * 100).sort_values()
print(missing_values_hits)
print('----------------------')
print(missing_values_sessions)
```

```python
len(df_sessions[(df_sessions['device_brand'] == 'Apple') & (df_sessions['device_os'].isnull()) & (df_sessions['device_category'] == 'mobile')])
```

```python
#let's assign "iOS" to all mobile device from Apple
df_sessions.loc[(df_sessions['device_category'] == 'mobile') & (df_sessions['device_brand'] == 'Apple') & (df_sessions['device_os'].isnull()), 'device_os'] = 'iOS'  
```

```python
#Check how many null values in mobile apple devices
len(df_sessions[(df_sessions['device_brand'] == 'Apple') & (df_sessions['device_os'].isnull()) & (df_sessions['device_category'] == 'mobile')])
```

```python
#Delete 'device_model' column from dataframe since there 99% missing values
del df_sessions['device_model']
```

```python
#delete 'utm_keyword' column from dataframe as well since there 58% missing values
del df_sessions['utm_keyword']
```

```python
len(df_sessions[(df_sessions['device_os'].isnull()) & (df_sessions['device_browser'] == 'Android Webview')])
```

```python
missing_values_sessions = ((df_sessions.isna().sum() / len(df_sessions)) * 100).sort_values()
print(missing_values_sessions)
```

```python
#Assign 'Android' to all rows in device_os where device_browser == 'Android Webview'
df_sessions.loc[(df_sessions['device_category'] == 'mobile') & (df_sessions['device_browser'] == 'Android Webview') & (df_sessions['device_os'].isnull()), 'device_os'] = 'Android'  
```

```python
#Assign 'Safari' to all rows in device_os where device_browser == 'Safari'
df_sessions.loc[(df_sessions['device_browser'] == 'Safari') & (df_sessions['device_os'].isnull()), 'device_os'] = 'iOS'  
```

```python
#Assign device_brand to "Apple" if device_os is "Macintosh"
df_sessions.loc[df_sessions['device_browser'] == 'Safari', 'device_brand'] = 'Apple'  
```

```python
df_sessions.loc[df_sessions['device_os'] == 'Macintosh', 'device_brand'] = 'Apple'
```

```python
df_sessions.loc[df_sessions['device_brand'] == 'Apple', 'device_os'] = 'iOS'
```

```python
df_sessions.loc[(df_sessions['device_brand'] == 'Google') & (df_sessions['device_category'] == 'mobile'), 'device_os'] = 'Android'
```

```python
#delete all rows where utm_adcontent is null
df_sessions = df_sessions.dropna(subset=['utm_adcontent']) 
df_sessions = df_sessions.dropna(subset=['utm_campaign']) 
df_sessions = df_sessions.dropna(subset=['utm_source']) 
```

```python
#set 'other' where missing value in device_brand
df_sessions.loc[df_sessions['device_brand'].isnull(), 'device_brand'] = 'other'
```

```python
#now we will set 'other'
df = df_sessions
df['device_category'].unique()
```
